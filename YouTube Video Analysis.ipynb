```python
data["title_length"] = data["title"].apply(lambda x: len(x))

fig, ax = plt.subplots()
_ = sns.distplot(data["title_length"], kde=False, rug=False, hist_kws={'alpha': 1}, ax=ax)
_ = ax.set(xlabel="Title Length", ylabel="No. of videos", xticks=range(0, 110, 10))
```

```python
sns.scatterplot(x = data['views'], y = data['title_length'])
```

```python
plt.figure(figsize = (10,10))
sns.heatmap(data.corr(), cbar = True, square = True, fmt = '.1f', annot = True, annot_kws = {'size':8}, cmap = 'Reds')
```

```python
title_words = list(data["title"].apply(lambda x: x.split()))
title_words = [x for y in title_words for x in y]
wc = wordcloud.WordCloud(width=1200, height=500, 
                         collocations=False, background_color="white", 
                         colormap="tab20b").generate(" ".join(title_words))
plt.figure(figsize=(15,10))
plt.imshow(wc, interpolation='bilinear')
_ = plt.axis("off")
```
